#### Plot dots and linear relationship

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
X = np.array([1.0, 2.0, 3, 4], dtype=np.float32)
Y = np.array([2.3, 3.4, 6.5, 6.8], dtype=np.float32)

# Create the plot
plt.scatter(X, Y, color='blue', label='Data points')  # Plot data points
plt.plot(X, 3*X, 'r--', label='y = 3x ')  # Plot the ideal line

# Add labels and title
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('Linear Relationship Plot')

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Set axis limits
plt.xlim(0, 5)
plt.ylim(0, 9)

# Show the plot
plt.show()
```

#### Calcuate Mean Squared Error when slope = 3
- w to reprsent the slope

```python
w= 3
y_pred = 3.0 * X  # Model: y = 2x
mse = np.mean((Y - y_pred) ** 2)
print(f"Mean Squared Error: {mse:.3f}")
```

#### Calcuate Mean Squared Error when slope range between [-5, 9.05] with 0.05 inteval
- w to reprsent the slope

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1.0, 2.0, 3, 4], dtype=np.float32)
Y = np.array([2.3, 3.4, 6.5, 6.8], dtype=np.float32)

# Generate weights from -5 to 5
w_values = np.arange(-5, 9.05, 0.05)

# Calculate MSE for each weight
y_preds = w_values[:, np.newaxis] * X
errors = Y - y_preds
mse_values = np.mean(errors**2, axis=1)

# Find optimal weight
min_idx = np.argmin(mse_values)
optimal_w = w_values[min_idx]
min_mse = mse_values[min_idx]

# Create plot
# plt.figure(figsize=(10, 6))
plt.plot(w_values, mse_values)
plt.scatter(optimal_w, min_mse, color='red', 
           label=f'Minimum MSE (w={optimal_w:.2f}, MSE={min_mse:.2f})')
plt.xlabel('Weight (w)', fontsize=12)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.title('MSE vs. Weight Parameter', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
